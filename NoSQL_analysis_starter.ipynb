### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
# query =

# Use count_documents to display the number of documents in the result

# Display the first document in the results using pprint


from pprint import pprint

# Define the query criteria
query = {
    'LocalAuthorityName': {'$regex':'London'},
    'RatingValue': {'$gte': 4}
}

# Find documents that match the query
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
num_docs = establishments.count_documents(query)
print("Number of documents:", num_docs)

# Check if any documents are returned
if num_docs > 0:
    # Display the first document in the results using pprint
    first_doc = results[0]
    pprint(first_doc)
else:
    print("No documents found.")


```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results


# Define the pipeline
pipeline = [
    {
        '$match': {
            'scores.Hygiene': 0
        }},{'$group': {'_id': '$LocalAuthorityName','count': {'$sum':1}}},
          {'$sort': {'count': -1}
    }
]

# Execute the pipeline
results = establishments.aggregate(pipeline)

# Rest of the code to print the results


# Get the estimated number of documents in the result
num_docs = establishments.estimated_document_count()
print("Number of documents:", num_docs)

# Print the first 10 results
for i, doc in enumerate(results):
    if i < 10:
        pprint(doc)
    else:
        break
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame

import pandas as pd

# Execute the pipeline
results = establishments.aggregate(pipeline)

# Convert the result to a DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
num_rows = len(df)
print("Number of rows:", num_rows)

# Display the first 10 rows of the DataFrame
print(df.head(10))

```
